```python
import ga
import numpy as np
```

```python
def objective(d): return np.abs(32 - d)
def decode(x): 
    x = ''.join(str(i) for i in x)
    return int(x, 2)
```

```python
n_iter = 200
n_bits = 4
n_pop = 100
r_cross = 0.9
bounds = 10
# r_mut = 1.0 / (float(n_bits) * len(bounds))
r_mut = .05

debug = False

settings = [n_bits, bounds, n_iter, n_pop, r_cross, r_mut, debug]
```

```python
best, score, best_scores, avg = ga.genetic_algorithm(objective, decode, *settings)
print('done')
print(score)
```
